# Using countVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(tokenizer=lambda doc1: doc1, lowercase=False)
X = vec.fit_transform(x).toarray()
print(X.shape)
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state =42)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2))
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion='entropy',max_depth=100)
dtree.fit(x_train,y_train)
y_pred=dtree.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA()
lda.fit_transform(x_train, y_train)
y_pred = lda.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn import svm
SVM = svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='scale', coef0=0.0, shrinking=True, probability=True, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None)
SVM.fit(x_train, y_train)
# predict the labels on validation dataset
y_pred = SVM.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
acc
```

```python
#Create adaboost classifer object
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=50,learning_rate=1)
model = abc.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(x_train,y_train)
y_pred = kn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
acc
```
